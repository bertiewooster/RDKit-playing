```python
from rdkit import Chem
from rdkit.Chem.MolStandardize import rdMolStandardize
from rdkit.Chem.Draw import MolsToGridImage

enumerator = rdMolStandardize.TautomerEnumerator()

smi1 = 'COc1ccc(C2C(C#N)=C3Sc4ccccc4N3C(=N)C2C#N)cc1'
mol1 = Chem.MolFromSmiles(smi1)
mol1
```

```python
tauts1 = enumerator.Enumerate(mol1)
MolsToGridImage(tauts1)
```

```python
smi1_tauts = [Chem.MolToSmiles(m) for m in tauts1]
```

```python
smi1_canon = Chem.MolToSmiles(mol1)
[smi1_canon == smi1_taut for smi1_taut in smi1_tauts]
```

```python
smi2 = 'COc1ccc(cc1)C1C(C#N)=C2Sc3ccccc3N2C(N)=C1C#N'
mol2 = Chem.MolFromSmiles(smi2)
tauts2 = enumerator.Enumerate(mol2)
MolsToGridImage(tauts2)
```

```python
smi2_tauts = [Chem.MolToSmiles(m) for m in tauts2]
[smi1 == smi2_taut for smi2_taut in smi2_tauts]
```

```python
smi2_canon = Chem.MolToSmiles(mol2)
smi2_canon
```

```python
[smi2_canon == smi2_taut for smi2_taut in smi2_tauts]
```
