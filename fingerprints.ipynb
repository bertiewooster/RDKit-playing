```python
from rdkit.Chem import AllChem
from rdkit import DataStructs
```

```python
fpgen = AllChem.GetMorganGenerator(radius=2)
```

```python
m1 = AllChem.MolFromSmiles('Cc1ccccc1')
m1
```

```python
fp1 = fpgen.GetSparseCountFingerprint(m1)
```

```python
m2 = AllChem.MolFromSmiles('Cc1ncccc1')
m2
```

```python
fp2 = fpgen.GetSparseCountFingerprint(m2)
```

```python
DataStructs.DiceSimilarity(fp1, fp2)
```

```python
ms = [AllChem.MolFromSmiles('CCOC'), AllChem.MolFromSmiles('CCO'),
AllChem.MolFromSmiles('COC')]
fps = [fpgen.GetFingerprint(x) for x in ms]
fpgenRDKit = AllChem.GetRDKitFPGenerator()
# fp1RDKit = fpgenRDKit.GetFingerprint(m1)
# fp2RDKit = fpgenRDKit.GetFingerprint(m2)
# DataStructs.TanimotoSimilarity(fp1RDKit, fp2RDKit)
DataStructs.TanimotoSimilarity(fps[0],fps[1])
```
